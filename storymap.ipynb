```python
import pandas as pd
import plotly.express as px

# read in the csv file
df = pd.read_csv('diabetes-prevalence.csv')

# filter for 2011 and 2021 data separately
df_2011 = df[df['Year']==2011]
df_2021 = df[df['Year']==2021]

# create graphs for both years
for df in [df_2011, df_2021]:
    fig = px.choropleth(df, locations='Entity',
                        locationmode='country names',
                        color='Diabetes prevalence (% of population ages 20 to 79)',
                        color_continuous_scale='Reds',
                        range_color=(0,25),
                        labels={'Diabetes_rate':'Diabetes Rate'},
                        title=f'World Diabetes Rates in {df["Year"].unique()[0]}'
                        )
    fig.show()
```

```python
high_ob = food[food.Obesity > food['Obesity'].mean()]
low_ob = food[food.Obesity <= food['Obesity'].mean()]
```

```python
fig = px.pie(values = high_ob[vegetal_features].mean().tolist(), names = vegetal_features,
             title='High obesity rate countries(vegetal_features)')
fig.show()

fig = px.pie(values = low_ob[vegetal_features].mean().tolist(), names = vegetal_features,
             title='Low obesity rate countries(vegetal_features)')
fig.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_round_data = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
```

```python
col = all_round_data.columns
all_round_data.columns
```

```python
col_isna = all_round_data.isnull().sum().to_list()
for count, i in enumerate(col_isna):
    print(count, '{:15}'.format(col[count]), i)
```

```python
all_round_data = all_round_data.rename(columns={'family_history_with_overweight': 'FHWO',
                           'NObeyesdad' : 'BMI',})
```

```python
object_cols = [col for col in all_round_data.columns if all_round_data[col].dtype == "object"]

good_label_cols = [col for col in object_cols if set(all_round_data[col]).issubset(set(all_round_data[col]))]


print('Categorical columns that will be ordinal encoded:', good_label_cols)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import OrdinalEncoder
```

```python
all_round_data_transform = all_round_data.copy()
ordinal_encoder = OrdinalEncoder()
all_round_data_transform[good_label_cols] = ordinal_encoder.fit_transform(all_round_data_transform[good_label_cols])
```

```python
dic_to_replace = {"BMI": {"Insufficient_Weight": 0,
                          "Normal_Weight": 1,
                         'Overweight_Level_I': 2,
                          'Overweight_Level_II': 3,
                          'Obesity_Type_I': 4,
                          'Obesity_Type_II': 5,
                          'Obesity_Type_III': 6,}
                         }
all_round_data.replace(dic_to_replace, inplace=True)
```

```python
corr = all_round_data_transform.corr()
```

```python
sns.set_theme(style="white")
f, ax = plt.subplots(figsize=(16, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.color_palette("YlOrBr", as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},
            annot = True, annot_kws={"size": 8})
plt.savefig('correlation.png', bbox_inches='tight',pad_inches=0)
```
