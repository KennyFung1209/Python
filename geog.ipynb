```python
fire = pd.read_csv('inpe_brazilian_amazon_fires_1999_2019.csv')
```

```python
fire
```

```python
fire_amazons = fire.loc[fire.state == 'AMAZONAS' ]
print(fire_amazons)
```

```python
#calculate sum of sales grouped by month
fire_amazons_yearly = fire_amazons.groupby(fire_amazons.year)['firespots'].sum()
print(fire_amazons_yearly)
```

```python

import matplotlib.pyplot as plt

# dataset
years = [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
Total_fireplots = [1048, 857, 1297, 10203, 10191, 8083, 15644, 11697, 11293, 6701, 9280, 12139, 5028, 9114, 6512, 9008, 13419, 11173, 11685, 11446, 12665]

# create a line graph
plt.plot(years, Total_fireplots)

# set the title
plt.title("Firespots in Amazons form 1999 to 2019")

# set the x and y labels
plt.xlabel("Year",fontsize=14)
plt.ylabel("Total Fireplots",fontsize=14)

plt.xlim([1999, 2019])
plt.ylim([0, 20000])
# show the graph
plt.xticks(np.arange(min(years), max(years)+1, 2.0))
plt.grid(True)
plt.show()

```
