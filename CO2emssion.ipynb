```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = files.upload()
```

```python
df = pd.read_csv('regression.csv') 
df
```

```python
from sklearn import linear_model
```

```python
x=df['Mean Global Surface temperature index'].values.reshape(-1,1)
x
```

```python
y=df['wheat price'].values.reshape(-1,1)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
```

```python
a = lm.intercept_
a
```

```python
b = lm.coef_
b
```

```python
y_prediksi = model.predict(x)
plt.scatter(x, y)
plt.plot(x, y_prediksi, c='r')
plt.xlabel('Mean Global Surface temperature index')
plt.ylabel('wheat price')
plt.title('Plot x vs y')
```

```python
MSE = mean_squared_error(y, y_prediksi)
MSE
```

```python
R2 = r2_score(y, y_prediksi)
R2
```
